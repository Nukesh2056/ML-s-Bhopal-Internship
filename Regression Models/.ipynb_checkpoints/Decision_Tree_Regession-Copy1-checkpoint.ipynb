```python
import pandas as pd 
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.multioutput import MultiOutputRegressor
```

```python
weather = pd.read_csv('ready_to_use_data.csv')
weather.keys()
```

```python
df = pd.DataFrame(weather)
df
```

```python
# Creating 2 columns each with next day min max for predicting next day min & max temperature.
df['Next_Max'] = df['Max.'].shift(-1)
df['Next_Min'] = df['Min.'].shift(-1)
df
```

```python
# Drop last row because it has NaN after shifting
df = df.dropna()
df.to_csv('ready_to_use_data_for_regression.csv', index=False)
```

```python
# sn.pairplot(df)
```

```python
x = df.drop(columns=['Next_Min' , 'Next_Max'])
y = df[['Next_Min', 'Next_Max']]

x_train , x_test , y_train , y_test = train_test_split(x , y , train_size=0.8 , random_state=42)

```

```python
model = MultiOutputRegressor(DecisionTreeRegressor(splitter='best' , max_depth=250 , min_samples_split = 4 , max_features = 'sqrt' , random_state=42 ))
model.fit(x_train , y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
# Plot only first 100 points
n = 100

plt.figure(figsize=(12, 5))

# Min Temp
plt.subplot(1, 2, 1)
plt.plot(y_test['Next_Min'].values[:n], label='Actual Min', marker='o')
plt.plot(y_predicted[:n, 0], label='Predicted Min', marker='x')
plt.title('Min Temp (First 100 samples)')
plt.xlabel('Sample')
plt.ylabel('Temp')
plt.legend()

# Max Temp
plt.subplot(1, 2, 2)
plt.plot(y_test['Next_Max'].values[:n], label='Actual Max', marker='o')
plt.plot(y_predicted[:n, 1], label='Predicted Max', marker='x')
plt.title('Max Temp (First 100 samples)')
plt.xlabel('Sample')
plt.ylabel('Temp')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

min_error = y_test['Next_Min'].values - y_predicted[:, 0]
max_error = y_test['Next_Max'].values - y_predicted[:, 1]

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(min_error, bins=30, kde=True, color='green')
plt.title('Min Temp Error Distribution')

plt.subplot(1, 2, 2)
sns.histplot(max_error, bins=30, kde=True, color='blue')
plt.title('Max Temp Error Distribution')

plt.tight_layout()
plt.show()
```

```python
mae = mean_absolute_error(y_test, y_predicted)
mse = mean_squared_error(y_test, y_predicted)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_predicted)
print("Results :")
print(f"MAE: {mae:.2f} °C")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f} °C")
print(f"R² Score: {r2:.2f}")

# Results : for parameters (splitter='best' , max_depth=250 , min_samples_split = 4) 
# MAE: 1.63 °C
# MSE: 11.63
# RMSE: 3.41 °C
# R² Score: 0.68

# Results : for parameters (splitter='best' , max_depth=250 , min_samples_split = 4 , max_features = 'sqrt') # this is the best results we have archived
# MAE: 1.64 °C
# MSE: 7.34
# RMSE: 2.71 °C
# R² Score: 0.80
```
