```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score , confusion_matrix
import numpy as numpy

weather = pd.read_csv('ready_to_use_data.csv')
```

```python
print(weather.keys())
# Index(['Max.', 'Min.', 'Morn', 'Eve', 'Rain', 'Rainy Days ', 'PE', 'year','month', 'day'],dtype='object')

data = weather
df = pd.DataFrame(data)
```

```python
df
```

```python
x = df.drop('Rainy Days ' , axis=1)
y = df['Rainy Days ']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , train_size=0.8)
```

```python
model = LogisticRegression(max_iter=1000) # default iteration is 100 , but model does not perform good in 100 interation so I've incresed the iteration 
model.fit(x_train , y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
cmetrix = confusion_matrix(y_test , y_predicted) 
cmetrix
```

```python
# visualizing it with seborn library 
plt.figure(figsize=(7,5))
sns.heatmap(cmetrix , annot = True , fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
print(f"Accuracy:, {accuracy_score(y_test, y_predicted)*100} % ")
```
