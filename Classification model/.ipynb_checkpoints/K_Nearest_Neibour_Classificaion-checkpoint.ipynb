```python
import pandas as pd 
import seaborn as sn

from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix , accuracy_score
from sklearn.neighbors import KNeighborsClassifier
weather = pd.read_csv('ready_to_use_data.csv')
```

```python
print(weather.keys())
# Index(['Max.', 'Min.', 'Morn', 'Eve', 'Rain', 'Rainy Days ', 'PE', 'year','month', 'day'],dtype='object')
data = weather
```

```python
data
```

```python
df = pd.DataFrame( data )
df
```

```python
x = df.drop('Rainy Days ' , axis=1)
y = df['Rainy Days ']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2 )
```

```python
# Applying the kNN model 
KNN_model = KNeighborsClassifier(n_neighbors=5 ,  leaf_size=30 , p=3)
KNN_model.fit(x_train , y_train)

# *, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=dict, n_jobs=None
```

```python
y_predicted = KNN_model.predict(x_test)
# y_predicted # this is predicted values by the KNN_model
```

```python
cmetrix = confusion_matrix(y_test , y_predicted) 
cmetrix
```

```python
# visualizing it with seborn library 
plt.figure(figsize=(7,5))
sn.heatmap(cmetrix , annot = True , fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
print(f"Model predictiion was {accuracy_score(y_test, y_predicted, normalize=True, sample_weight=None)*100} % Accurate")
```
