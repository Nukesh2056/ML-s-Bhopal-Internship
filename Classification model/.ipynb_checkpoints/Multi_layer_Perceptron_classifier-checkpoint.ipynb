```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix ,accuracy_score 
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
import seaborn as sn
weather = pd.read_csv('ready_to_use_data.csv')
```

```python
df = pd.DataFrame( weather )
df
```

```python
x = df.drop('Rainy Days ' , axis=1)
y = df['Rainy Days ']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
model = MLPClassifier(hidden_layer_sizes=100 , activation='logistic' ) # without hidden layer paramemter , accuracy is 54% 
model.fit(x_train , y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
cmetrix = confusion_matrix(y_test , y_predicted) 
cmetrix
```

```python
# visualizing it with seborn library 
plt.figure(figsize=(7,5))
sn.heatmap(cmetrix , annot = True , fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
print(f"Accuracy: {accuracy_score(y_test, y_predicted)*100} % ")
```
