```python
# NOTE : Feature scaling is very Crucial in SVM 

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix ,accuracy_score 
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import seaborn as sn
weather = pd.read_csv('ready_to_use_data.csv')
```

```python
df = pd.DataFrame( weather )
df
```

```python
x = df.drop('Rainy Days ' , axis=1)
y = df['Rainy Days ']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2 , random_state=42)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = SVC(kernel="linear") # the kernel parameter is  crucial to get 99% accuracy , it must be linear
model.fit(x_train_scaled,y_train)
```

```python
y_predicted = model.predict(x_test_scaled)
y_predicted
```

```python
cmetrix = confusion_matrix(y_test , y_predicted) 
cmetrix
```

```python
# visualizing it with seborn library 
plt.figure(figsize=(7,5))
sn.heatmap(cmetrix , annot = True , fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
print(f"Accuracy Score : {accuracy_score(y_test , y_predicted)*100}")
```
